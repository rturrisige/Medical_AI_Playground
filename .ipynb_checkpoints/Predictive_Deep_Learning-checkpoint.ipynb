**Useful Functions**

Let's write some useful function that will allow us to easily load the images and visualize the results.

```python
def load_data(img_dir, img_dim=None, nmax=None):
    """
    It reads and returns the image files in the subfolders contained in img_dir

    Arguments
    ---------
    img_dir: path to the directory containing the dataset
    img_dim: if not None, the images are resized to img_dim
    nmax: if not None, it corresponds to the maximum number of images per class to load

    Returns
    ---------
    X : tensor nmax*xdim*ydim of images
    y : numerical labels associated to X
    label : number of different labels
    label_names : list of label names
    """
    label = 0
    label_names = []
    X = []
    y=[]
    for dirname in os.listdir(img_dir):
        if not dirname.startswith('.'):  # does not consider hidden folders
            label_names.append(dirname)
            data_path = os.path.join(img_dir + "/" + dirname,'*g')
            files = glob.glob(data_path)  # create the list of files contained in the folder
            n=1
            for f1 in files:
                if nmax:
                    if n>nmax :
                        print('The folder', dirname, '(Label: ', label, ') contains', n-1,'images')
                        break
                img = cv2.imread(f1) # it reads the image file
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Conversion to RGB colors
                if img_dim:
                    img = cv2.resize(img, img_dim) # Resize the
                X.append(np.array(img))
                y.append(label)
                n=n+1
            label = label+1
    X = np.array(X)
    y = np.array(y)
    return X,y, label, label_names


def plot_scores(training):
    '''
    It takes the trained model and plots a figure with 2 subplots:
    1 training and validation loss
    2 training and validation accuracy.

    Arguments
    ---------
    training: trained model, output of model.fit

    '''
    loss = training.history['loss']
    accuracy = training.history['accuracy']
    val_accuracy = training.history['val_accuracy']
    val_loss = training.history['val_loss']
    epochs = range(len(accuracy))
    plt.figure()
    plt.subplot(1,2,1)
    plt.title('Loss')
    plt.plot(epochs, loss, 'b', label='train')
    plt.plot(epochs, val_loss, 'r', label='val')
    plt.legend()
    plt.subplot(1,2,2)
    plt.title('Accuracy')
    plt.plot(epochs, accuracy, 'b', label='train')
    plt.plot(epochs, val_accuracy, 'r', label='val')
    plt.legend()
    plt.show()
```

## **Dataset loading a pre-processing**

```python
# Define parameters:
input_dim = (126, 126) # for computational reasons we fix the dimension of the image to 126
```

```python
# Load training data
X,y, n_classes, classes = load_data(path + "/dataset/images", input_dim, nmax=1200)
print('Total size of the dataset:', X.shape[0])
```

```python
# Print data information
print('Number of classes:', n_classes)
print('Classes:', classes)
print('')
print('Dataset size:')
print('Train X:', X.shape, 'y:', y.shape)
print('')
print('Example of y data:', y[0])
```

### Dataset split: training, validation and testing set

```python
# Create the Validation set
n_val = X.shape[0]/100 * 20  # corresponds to the 20% of the whole dataset size
n_val_class = int(n_val/n_classes)
X_val = np.concatenate([X[:n_val_class], X[:-n_val_class]], 0)
y_val = np.concatenate([y[:n_val_class], y[:-n_val_class]], 0)
print('N image in validation per class:', n_val_class)
```

```python
n_test = X.shape[0]/100 * 20  # corresponds to the 20% of the whole dataset  size
n_test_class = int(n_test/n_classes)
n_val_test = n_test_class + n_val_class
X_test = np.concatenate([X[n_val_class:n_val_test], X[-n_val_test:-n_val_class]], 0)
y_test = np.concatenate([y[n_val_class:n_val_test], y[-n_val_test:-n_val_class]], 0)
print('N image in validation per class:', n_test_class)
```

```python
X_train = X[n_val_test:-n_val_test]
y_train = y[n_val_test:-n_val_test]
```

### **Data standardization**

```python
# Data standardization
X_train = X_train / 255
X_val = X_val / 255
X_test = X_test / 255
```

```python
# Labels conversion to one-hot vectors
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
y_test = to_categorical(y_test)
```

```python
# Print final dataset information

print('Train: X', X_train.shape, 'y:', y_train.shape)
print('Validation X:', X_val.shape, 'y:', y_val.shape)
print('Test X:', X_test.shape, 'y:', y_test.shape)
print('')
print('Example of y:', y_test[0])
```

## **CNN: Model definition and training**

```python
# Define the model
model_cnn = Sequential()
model_cnn.add(Conv2D(32, (3, 3), input_shape=(input_dim[0], input_dim[1], 3), activation='relu'))
model_cnn.add(Flatten())
model_cnn.add(Dense(n_classes, activation='softmax'))

model_cnn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

es = EarlyStopping(monitor='val_loss', mode='min', patience=5, verbose=1)
```

```python
# Model training
train_cnn = model_cnn.fit(X_train,
                          y_train,
                          validation_data=(X_val, y_val),
                          epochs=20,
                          batch_size=200,
                          verbose=1,
                          callbacks=[es])

model_cnn.summary()
```

```python
# Plot training results
plot_scores(train_cnn)
```

### **Model Evaluation**

```python
# Evaluate the model on the testing set

scores = model_cnn.evaluate(X_test, y_test, verbose=0)
print("Score : %.2f%%" % (scores[1]*100))
```

```python
prediction_test = model_cnn.predict(X_test)
cm = confusion_matrix(np.argmax(y_test, 1), np.argmax(prediction_test,1))

ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt="d")

ax.set_title('Confusion Matrix\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(classes)
ax.yaxis.set_ticklabels(classes)

## Display the visualization of the Confusion Matrix.
plt.show()
```

## Grad-CAM for interpretability
The Grad-Cam algorithm is a tool to inspect your model and looks at the features that ended up activated in the final activation map. This method has been firstly introduced in this [work](https://arxiv.org/pdf/1610.02391.pdf). You can also find an hands-on explanation [here](https://medium.com/@stepanulyanin/implementing-grad-cam-in-pytorch-ea0937c31e82).

```python
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # First, we create a model that maps the input image to the activations
    # of the last conv layer as well as the output predictions
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )

    # Then, we compute the gradient of the top predicted class for our input image
    # with respect to the activations of the last conv layer
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    # This is the gradient of the output neuron (top predicted or chosen)
    # with regard to the output feature map of the last conv layer
    grads = tape.gradient(class_channel, last_conv_layer_output)

    # This is a vector where each entry is the mean intensity of the gradient
    # over a specific feature map channel
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # We multiply each channel in the feature map array
    # by "how important this channel is" with regard to the top predicted class
    # then sum all the channels to obtain the heatmap class activation
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    print(heatmap.shape)
    heatmap = tf.squeeze(heatmap)
    print(heatmap.shape)
    # For visualization purpose, we will also normalize the heatmap between 0 & 1
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def visualize_img_and_grad(img, heatmap):
  # Display heatmap
  plt.figure(figsize=(10,10))
  plt.subplot(1,3,1)
  plt.imshow(img)
  plt.title("Original Image")

  plt.subplot(1,3,2)
  plt.imshow(heatmap)
  plt.title("Grad-CAM")

  plt.subplot(1,3,3)
  plt.imshow(img)
  plt.imshow(heatmap , alpha = 0.5, cmap=plt.cm.RdBu_r)  # FILL IN THE VALUE
  plt.title("Overlayed Grad-CAM")
  plt.show()
```

```python
model_cnn.layers[-1].activation = None

for i in [0, 1, 2,  -1, -2, -3]:
  img = X_val[i, :, :, :]
  heatmap = make_gradcam_heatmap(img[None, :, :, :], model_cnn, "conv2d")
  visualize_img_and_grad(img, heatmap)
```
