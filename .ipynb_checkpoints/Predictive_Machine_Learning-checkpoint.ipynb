### 1.2 Exploiting all variables simultaneously
The dimensionality of our dataset is 9. In order to actually *see* our data points, let's project them in a two-dimensional space.
To this aim, we can use [`PCA`](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) and [`TSNE`](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html). Let's first transform the `DataFrame` object into `numpy` array.

```python
numpy_data = data.to_numpy()
X, y = numpy_data[:, :-1], numpy_data[:, -1]
```

```python
X_PCA_reduced = PCA(n_components=2).fit_transform(X,y)
print("Reduced X shape = {}".format(X_PCA_reduced.shape))

for label in np.unique(y):
    idx = np.where(y == label)[0]
    plt.plot(X_PCA_reduced[idx, 0], X_PCA_reduced[idx, 1], 'o', label=int(label))
plt.xlabel("$x_0$")
plt.ylabel("$x_1$")
plt.legend();
```

```python
X_TSNE_reduced = TSNE(n_components=2, init='pca').fit_transform(X,y)#PCA(n_components=2).fit_transform(X, y)
print("Reduced X shape = {}".format(X_TSNE_reduced.shape))

for label in np.unique(y):
    idx = np.where(y == label)[0]
    plt.plot(X_TSNE_reduced[idx, 0], X_TSNE_reduced[idx, 1], 'o', label=int(label))
plt.xlabel("$x_0$")
plt.ylabel("$x_1$")
plt.legend();
```

## Part 2: Classification model

In our learning problem we have the dataset (x, y) and we want to find a function   such that (). Specifically, this is a binary classification problem as $y \in \{0,1\}$.


### STEP 1: Input standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

### STEP 2: SPLIT IN TRAINING AND TESTING SETS

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=123)
print('Shape of {} is {}'.format("x_train",x_train.shape )) 
print('Shape of {} is {}'.format("y_train",y_train.shape))
print('')
print('Shape of {} is {}'.format("x_test",x_test.shape))
print('Shape of {} is {}'.format("y_test",y_test.shape))
```

```python
from collections import Counter 
print('Trainin set:')
print('Total number of samples:', y_train.shape[0])
train_counter = Counter(y_train)  
for k, v in train_counter.items():
  print('Label', k, ' - N. of samples:', v)
```

```python
print('Testing set:')
print('Total number of samples:', y_train.shape[0])
test_counter = Counter(y_test)  
for k, v in test_counter.items():
  print('Label', k, ' - N. of samples:', v)
```

### STEP 3: TRAIN A MODEL

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(x_train,y_train)
```

### STEP 4: EVALUATE THE MODEL ON THE TESTING SET

```python
y_pred = lg.predict(x_test)
print(y_pred)
```

## How can we assess the model performance?
1. [Accuracy](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html?highlight=accuracy#sklearn.metrics.accuracy_score): rate of correct classifications


```python
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:', accuracy)
```

We need to introduce the concepts of:
* True positive (TP). Ex: a patient correctly classified as patient
* False positive (FP). Ex: healthy subject missclassified as patient
* True negative (TN). Ex: healthy subject correctly classified as heatlhy
* False negative (FN). Ex: patient missclassified as healthy.

2. [Confusion matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html?highlight=confusion%20matrix#sklearn.metrics.confusion_matrix): plots the TP, FP, TN and FN.
image.png


```python
cm2 = confusion_matrix(y_test,y_pred, labels=[1, 0])
img = sns.heatmap(cm2,annot=True,cmap='Blues')
img.set(xlabel="True Labels", ylabel="Prediction")
img.xaxis.set_ticklabels(['Diabetes', 'Healthy']);
img.yaxis.set_ticklabels(['Diabetes', 'Healthy']);

plt.title('Confusion matrix:')
plt.show()
```

3. Precision (specificity): portion of positive identifications correctly classified, i.e. TP/(TP+FP)
4. Recall (sensitivity): portion of actual positive correctly classified, i.e. TP/(TP+FN)
5. [F1 score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html?highlight=f1%20score#sklearn.metrics.f1_score): F1 = 2 * (precision * recall) / (precision + recall) 

```python
f1 = f1_score(y_test,y_pred)
print('F1 score:', f1)
```

```python
print(classification_report(y_test,y_pred))
```

- `support` tells us how many samples are in the dataset (e.g., 96 samples with 0 label).

- `macro avg` is the average of the `f1-score` measures computed for each label

- `weighted avg` is weighted average of the `f1-score` measures computed for each label (i.e., the `f1-score` measures multiplied by their support and divided by the total number of samples)


6. [ROC curve](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.plot_roc_curve.html?highlight=roc%20curve#sklearn.metrics.plot_roc_curve)(receiver operating characteristic curve): is a graph showing the performance of a classification model at all classification thresholds. This curve plots two parameters
* True Positive Rate (TPR) - a synonym for recall - defined as TP/(TP+FN)
* False Positive Rate (FPR) defined as FP/(FP+FN). This corresponds to 1 - precision.

An ROC curve plots TPR vs. FPR at different classification thresholds. Lowering the classification threshold classifies more items as positive, thus increasing both False Positives and True Positives. The following figure shows a typical ROC curve.



```python
FPR, TPR, threshold = roc_curve(y_test,y_pred) #use roc_curve function from sklearn library
area_under_curve = auc(FPR, TPR) #use auc function from sklearn library

plt.plot([0, 1], [0, 1], 'k--')
plt.plot(FPR, TPR, label='AUC = {:.3f}'.format(area_under_curve))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
```

AUC (Area under the curve):
* AUC=0.5 not informative;
* 0.5 < AUC 0.7 low accurate;
* 0.7 < AUC  0.9 moderately accurate;
* 0.9 < AUC <1.0 highly accurate;
* AUC = 1 perfect.
