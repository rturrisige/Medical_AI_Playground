# Part 1: Data Cleaning
Before working with data, we first need to have a look at the it and check if it contains some errors. The first thing to check is if there are features containing the same value for all samples: this features are not informative and can be discard. Then, we will check if there are NaN values corresponding to corrupted information.

### **1.1 Discard useless features**
Constant features do not bring information as they do not differentiate among classes.

```python
all_features = data.columns
useful_features = []
useless_features = []
for feat in all_features:
  if all(element == data[feat][0] for element in data[feat]):
    useless_features.append(feat) 
  else:
    useful_features.append(feat)

print('Tot number of features:', len(all_features))
print('Number of useless features:', len(useless_features))
print('Number of useful features:', len(useful_features))
print('')
print('Example of useless feature:', useless_features[0])
print(data[useless_features[0]])
```

```python
# Let's use only the useful features
data = data[useful_features]
print(data.shape)
```

### 1.2 discard corrupted data
Then, we need to check if data contains NaN values. This is a common issue when we upload a table in which cells are not filled.

```python
column_with_nan = []
for feat in useful_features:
  if(np.isnan(data[feat]).any()):
    print(feat, "contains NaN values")
    column_with_nan.append(feat) 

print('The input matrix contains', len(column_with_nan), 'columns with NaN values')
```

We can give a look to the resulting data by showing some statistics about it. We can use the `describe()` method.

```python
data.describe()
```

# Part 2: Exploratory data analysis

Before building the models, it is always useful to perform the exploratory data analysis.

Let's convert `DataFrame` objects into `numpy` arrays. We can do that using the method `to_numpy`. Then, we have to convert the string labels into numbers. 

```python
X, categorical_y = data.to_numpy(), labels.to_numpy()
print(X.shape, categorical_y.shape)
```

Our labels are strings, which are not treatable objects. We need to transform them into numbers. We can do this by using the dictionary object.

```python
conversion={'BRCA':0, 'COAD':1, 'KIRC':2, 'LUAD':3, 'PRAD':4}
y = np.empty([categorical_y.shape[0]])
for i in range(categorical_y.shape[0]):
  y[i] = conversion[categorical_y[i, 0]]

print('Conversion example:')
print('String labels:', categorical_y[:3].T)
print('Converted labels:', y[:3])
```

**Data Visualization** 

The dimensionality of our dataset is too big to be visualized!

In order to actually *see* our data points, let's project them in a two-dimensional space.
To this aim, we can use [`PCA`](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html).

```python
X_2D = PCA(n_components=2).fit_transform(X)
print("Reduced X shape = {}".format(X_2D.shape))
```

```python
# 2D visualization
for l in np.unique(labels):
    idx = np.where(labels == l)[0]
    plt.plot(X_2D[idx, 0], X_2D[idx, 1], 'o', label=l)
plt.xlabel("$x_0$")
plt.ylabel("$x_1$")
plt.legend()
plt.title('2D Data (PCA)')
```

### **Correlation analysis** 
We can exploit the correlation between pairs of variables by using `corrcoeff()` the method. 

```python
# compute the correlation matrix among features
cor_mat = np.corrcoef(X.T)
```

```python
# check which features are highly correlated (more than 90% and more than 98%)
very_high_correlated_feats = []
high_correlated_feats = []
for i in range(cor_mat.shape[1]):
  for j in range(i+1, cor_mat.shape[0]):
    if cor_mat[i, j] >= 0.98:
      very_high_correlated_feats.append([i,j])
    elif cor_mat[i, j] < 0.98 and cor_mat[i, j] > 0.90:
      high_correlated_feats.append([i,j])

print('Number of pairs of very high correlated features (> or = 98%):', len(very_high_correlated_feats))
print('Number of pairs of high correlated features (> 90%):', len(high_correlated_feats))
```

### Highly correlated features may bring the same information and, hence, it may not be necessary to keep all of them. 

```python
# discard one of the two paired correlated features
features_to_discard = []
for pairs in very_high_correlated_feats + high_correlated_feats:
  features_to_discard.append(pairs[1])

X = np.delete(X, features_to_discard, 1)
print('Number of removed features:', len(features_to_discard))
print('Final shape of the input matrix X:', X.shape)

# take the names of the kept features
feature_names = [data.columns[i] for i in range(data.columns.shape[0]) if i not in features_to_discard]
# delete information we do not need anymore to save memory
del cor_mat
```

# Part 3: Feature selection
We will perform feature selection using **Recursive Feature Elimination (RFE)**. In this step, feature selection is based on the outcome, so simple models are built and those features that contribute the least to the model are removed recursively. For this purpose, we first need to split the dataset into training and testing. 

Are the classes in the dataset balanced? We can check it by using `Counter` method.


```python
from collections import Counter
c = Counter(y)
for i in list(conversion.values()):
  l = list(conversion.keys())[i]
  print('Label', l, 'has', c[i], 'samples.')
```

As the dataset is not balanced, we cannot split it randomly as we would risk to have the smallest class to be not represented in the learning or testing sets. We can use a function that splits the dataset by keeping the class proportions.

```python
from sklearn.model_selection import StratifiedShuffleSplit
s = StratifiedShuffleSplit(n_splits=2, test_size=0.2, random_state=0)
learning_index, testing_index = next(s.split(X, y))
X_learn, y_learn = X[learning_index, :], y[learning_index]
X_test, y_test= X[testing_index, :], y[testing_index]
print('N. of samples in the learning set:', X_learn.shape[0])
print('N. of samples in the testing set:', X_test.shape[0])
```

<b>How many features do we need to select with RFE?
Time to discuss pros and cons of using many and a little features.</b>  
  
There are some rules of thumb on how many features we need in the end:  
* $int(\frac{N_{samples}}{10})$ (Abu-Mostafa, Y. S., Magdon-Ismail, M., & Lin, H. T. (2012). Learning from data (Vol. 4, p. 4). New York: AMLBook.)  



* $\sqrt{N_{samples}}$ (Hua, J., Xiong, Z., Lowey, J., Suh, E., & Dougherty, E. R. (2005). Optimal number of features as a function of sample size for various classification rules. Bioinformatics, 21(8), 1509-1515.)

```python
print ('Nmber of samples in training dataset: ', X_train.shape[0])
print ('Number of features to select according to Abu-Mostafa: ', int(X_train.shape[0]/10))
print ('Number of features to select according to Hua: ', int(X_train.shape[0]**0.5))
```

Let's go for the lower number of features since our dataset is not large. Below we implement Recursive Feature Elimination, RFE (https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html), based on Support Vector machine Classifier, SVC (https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html).

Which number of features will you chose?

```python
from sklearn.feature_selection import RFE
from sklearn.svm import SVC

N_feat_to_select = 23
estimator = SVC(kernel='linear')
selector = RFE(estimator, n_features_to_select=N_feat_to_select, step=100)
selector = selector.fit(X_train, y_train)
pred = selector.predict(X_test)
print('Testing accuracy:', accuracy_score(pred, y_test))
```

Let's investigate which are the most important features. We can use the `get.support()` method to return the important features. Specifically, it returns a matrix of True and False, where True corresponds to the selected features. We can also use the `ranking_` attribute rankes the feature importance (rank=1 for selected feature).

```python
support = selector.get_support()
important_features = np.where(support==True)[0]
important_features_name = [feature_names[i] for i in important_features]
print('The most important features are:')
for f in important_features_name:
  print(f)
```

```python
ranks = selector.ranking_
print('The rank varies from {} to {}'.format(min(ranks), max(ranks)))
```

We can transform the rank into weights, attributing the highest weight to the smallest rank. We can use the formula w = max(ranks) - rank + 1. We can also normalise the weights between 0 and 1 for a better reading.

```python
def from_rank_to_weight(rank):
  max = np.max(rank)
  w = [max - r + 1 for r in rank]
  m, M = np.min(w), np.max(w)
  return (w - m) / (M - m)

W = from_rank_to_weight(ranks)
```

```python
print(important_features)
```

We can plot the feature weights by using the bar plots.

```python
plt.figure(figsize=(20, 5))
plt.imshow(W[:, None].T, aspect='auto', cmap="viridis")
plt.colorbar()
plt.yticks([])
```

# Part 3: Models comparison
Are those really the important features?
We can compare the performance models trained on different input:
* The input matrix with features selected by RFE;
* The input matrix with features selected randomly.

```python
# Predictive model with RFE selected features:
from sklearn.svm import SVC
model_1 = SVC()
model_1.fit(X_train[:, important_features], y_train)
test_prediction_1 = model_1.predict(X_test[:, important_features])
model1_acc = accuracy_score(y_test, test_prediction_1)
```

```python
import random
# select random features:
indices = [i for i in range(X.shape[1])]
random.shuffle(indices)
random_selected_features = indices[:N_feat_to_select]

# Predictive model with random selected features:
model_2 = SVC()
model_2.fit(X_train[:, random_selected_features], y_train)
test_prediction_2 = model_2.predict(X_test[:, random_selected_features])
model2_acc = accuracy_score(y_test, test_prediction_2)
```

```python
print('Testing accuracy:')
print('Selected features:', model1_acc)
print('Random selected features:', model2_acc)
```
